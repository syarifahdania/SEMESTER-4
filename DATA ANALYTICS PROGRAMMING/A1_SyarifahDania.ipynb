#### QUESTION 2: Read and parse data from the```"ride.csv"```. Complete the given code (8 marks)

````python
import csv

def load_ride_data(filename):
    data = []
    with open(filename, mode='r') as file:
        reader = csv.DictReader(file) 
        #read and parse data iteratively  (5 marks)
        for row in reader:
            data.append(row)
    return data

# Write code(s) to load data from actual CSV file (2 marks)
rides = load_ride_data("ride.csv")

# Write code(s) to preview data (1 mark)
for ride in rides:
    print(ride)
````

#### QUESTION 3: Define a function to compute average fare by city and call the function by passing the data (10 marks)

````python
# Write your function definition here to compute average fare by city (8 marks)
def averageFareByCity(data):
    city_fares = {}
    city_counts = {}

    for ride in data:
        city = ride['city']
        fare = float(ride['fare'])

        if city not in city_fares:
            city_fares[city] = fare
            city_counts[city] = 1
        else:
            city_fares[city] += fare
            city_counts[city] += 1

    average_fares = {}
    for city in city_fares:
        average_fares[city] = round(city_fares[city] / city_counts[city] ,2)

    return average_fares


# Write the code(s) to call your function and pass the data (2 marks)
average_fares = averageFareByCity(rides)

for city, avg_fare in average_fares.items():
    print(f"{city}: ${avg_fare}")
````

#### QUESTION 4: Define a function to summarize driver information, call the function by passing data ad display the summary (10 marks)
#### ```driver_name, num_of_rides, total_distance, total_fare```

````python

# Write your function definition here (7 marks)
def driver_info(data):
    summary = {}

    for ride in data:
        driver = ride['driver_name']
        distance = float(ride['distance'])
        fare = float(ride['fare'])

        if driver not in summary:
            summary[driver] = {
                'num_of_rides': 1,
                'total_distance':distance,
                'total_fare': fare
        }
        else:
            summary[driver]['num_of_rides'] += 1
            summary[driver]['total_distance'] += distance
            summary[driver]['total_fare'] += fare

    return summary


# Write the code(s) to call your function and pass the data (2 marks)
driver_summary = driver_info(rides)

# Write the code(s) to display driver's summary (1 mark)
print("Driver Summary:")
print("driver_name , num_of_rides , total_distance , total_fare")
for driver, info in driver_summary.items():
    print(f"{driver} , {info['num_of_rides']} , {info['total_distance']} , {info['total_fare']}")

````

#### QUESTION 5: Implement a function ```analyze_peak_slice(data, start_idx, end_idx)``` that: (10 marks)

    1.Accepts the ride data list and two indices: start_idx and end_idx.
    2.Extracts a slice of rides from start_idx = 2 to end_idx = 6 (inclusive).
    3.Calculates and returns:
        i)  Total number of rides in that slice
        ii) Average fare
        iii)List of unique drivers involved
        
    Sample output:
 
   ```Total_rides: 5```<br>
   ```Average_fare: 18.45```<br>
   ```Drivers: Ravi, Ali, Mei``` 

````python
# Write your function definition here that accepts ride data, start index and end index, slice the list, 
# calculate and returns the values in (3)   (8 marks)
def analyze_peak_slice(data, start_idx, end_idx):
    slicedRides = data[start_idx:end_idx + 1]
    totalRides = len(slicedRides)
    totalFare = sum(float(ride['fare']) for ride in slicedRides)
    averageFare = round(totalFare / totalRides, 2)
    uniqueDrivers = list(set(ride['driver_name'] for ride in slicedRides))

    return totalRides, averageFare, uniqueDrivers

# Write the code(s) to call your function and pass the ride data, start index and end index (2 marks)
total_rides, avg_fare, drivers = analyze_peak_slice(rides, 2, 6)

print(f"Total_rides: {total_rides}")
print(f"Average_fare: {avg_fare}")
print("Drivers:",drivers)
````

#### QUESTION 7: Write a function to calculate average duration by cities and use lambda function to sort cities by average duration (10 marks)

````python
# Write your function definition here for average duration and sort cities (8 marks)
# function to calculate average duration and sort cities using lambda
def average_duration_by_city(data):
    from collections import defaultdict

    # 1)group durations by city
    city_duration = defaultdict(list)
    for city, duration in data:
        city_duration[city].append(duration)

    # 2)calculate average for each city
    avg_by_city = {city: sum(durations) / len(durations) for city, durations in city_duration.items()}

    # 3)sort cities by average duration using lambda
    sorted_avg = sorted(avg_by_city.items(), key=lambda x: x[1])

    return sorted_avg


# Write the code(s) to call your function and pass the data (2 marks)
# sample input data: (city, duration in minutes)
ride_data = [
    ("Kuala Lumpur", 30),
    ("Kuala Lumpur", 50),
    ("Johor Bahru", 20),
    ("Johor Bahru", 25),
    ("Penang", 45),
    ("Penang", 35)
]

# calling the function
sorted_cities = average_duration_by_city(ride_data)

# displaying the result
print("Cities sorted by average ride duration:")
for city, avg_duration in sorted_cities:
    print(f"{city}: {avg_duration:.2f} minutes")

````

#### QUESTION 8: Define a function to write driver summary to a text file ```driver_summary.txt``` with the following information:

````python
# Write your function definition here for writing driver summary into a text file (8 marks)
# function to write driver summary into a text file
def write_driver_summary(driver_data):
    with open("driver_summary.txt", "w") as file:
        file.write("Driver, Total Rides, Total Distance, Total Fare\n")
        for driver, total_rides, total_distance, total_fare in driver_data:
            file.write(f"{driver}, {total_rides}, {total_distance}, {total_fare}\n")









# Write the code(s) to call your function and pass the driver summary data (2 marks)
# sample drivers' data: (driver, total rides, total distance, total fare)
drivers = [
    ("Ali", 12, 130.5, 350.75),
    ("Ravi", 9, 95.2, 275.60),
    ("Sara", 15, 150.0, 420.90)
]

# code to call the function
write_driver_summary(drivers)

print("Driver summary has been written to 'driver_summary.txt'")


````
